```python
cyber['DATA SENSITIVITY'] = cyber['DATA SENSITIVITY'].replace({'financial':'4','oops!': '1', 'y':'2'  })
```

```python
cyber['DATA SENSITIVITY'] = cyber['DATA SENSITIVITY'].fillna(0).astype(int)
```

```python
cyber['cyber_date'] = pd.to_datetime(
    cyber['story'].apply(lambda x: x[:8]).str.upper(),
    format='%b %Y',
    yearfirst=False,errors='coerce')
```

```python
cyber['year'] = cyber['cyber_date'].apply(lambda x: x.year).fillna(0).astype(int)
cyber['month'] = cyber['cyber_date'].apply(lambda x: x.month).fillna(0).astype(int)
cyber['key'] = cyber['Ticker'].astype(str)+cyber['year'].astype(str)+cyber['month'].astype(str)
```

```python
stocks['date'] = pd.to_datetime(stocks.date)
```

```python
data = stocks[stocks['symbol'] == 'LNKD']
data = data.sort_values(by=['date'], ascending=True)
data.index = pd.DatetimeIndex(data['date'], name='date_')
data = data.resample('M').first()
data['returns'] = data['price']/data['price'].shift(1)
data['year'] = data['date'].apply(lambda x: x.year)
data['month'] = data['date'].apply(lambda x: x.month)
data = data.fillna(0)
data['year'] = data['year'].astype(int)
data['month'] = data['month'].astype(int)
data['d_key'] = data['symbol'].astype(str)+data['year'].astype(str)+data['month'].astype(str)
```

```python

metric = 'price'
res = pd.DataFrame()
for ticker in set(stocks['symbol']):
    data = stocks[stocks['symbol'] == ticker]
    data = data.sort_values(by=['date'], ascending=True)
    data.index = pd.DatetimeIndex(data['date'], name='date_')
    data = data.resample('M').first()
    data['returns'] = data['price']/data['price'].shift(1)
    data['year'] = data['date'].apply(lambda x: x.year)
    data['month'] = data['date'].apply(lambda x: x.month)
    data = data.fillna(0)
    data['year'] = data['year'].astype(int)
    data['month'] = data['month'].astype(int)
    data['d_key'] = data['symbol'].astype(str)+data['year'].astype(str)+data['month'].astype(str)
    d = data.merge(cyber, left_on=data.d_key, right_on=cyber.key, how='left')
    d = d.fillna(0)
    d['data breach'] = np.where(d['Entity']!=0,d[metric],0)
    d = d[['date', 'price', 'returns', 'symbol', 'volume', 'data breach']]
    d = d.iloc[1:]
    res = res.append(d)

res.to_csv('cyber_breaches_and_Stock_price.csv')
```

```python
def plot(ticker='YHOO', metric='price'):
    pyplot.show()
    fig = plt.figure(figsize=(15,9))
    data = stocks[stocks['symbol'] == ticker]
    data = data.sort_values(by=['date'], ascending=True)
    data.index = pd.DatetimeIndex(data['date'], name='date_')
    data['vol_sum'] = data['volume'].cumsum()
    data = data.resample('M').first()
    data['returns'] = data['price']/data['price'].shift(1)
    data['vol'] = data['vol_sum'] - data['vol_sum'].shift(1)
    data['year'] = data['date'].apply(lambda x: x.year)
    data['month'] = data['date'].apply(lambda x: x.month)
    data['d_key'] = data['symbol'].astype(str)+data['year'].astype(str)+data['month'].astype(str)
    d = data.merge(cyber, left_on=data.d_key, right_on=cyber.key, how='left')
    d = d.fillna(0)
    d['data breach'] = np.where(d['Entity']!=0,d[metric],-100)
    d = d[['date', 'price', 'returns', 'symbol', 'vol', 'data breach', 'DATA SENSITIVITY']]
    d = d.iloc[1:]
    plt.plot(d['date'], d[metric])
    plt.plot(d['date'], d['vol'], 'g-')
   # plt.scatter(d['date'].tolist(), d['data breach'], c=d['DATA SENSITIVITY'], s=d['DATA SENSITIVITY']*100, cmap = 'Reds')
    plt.scatter(d['date'].tolist(), d['data breach'],c='Red', s=d['DATA SENSITIVITY']*100)
    plt.ylim((0,d[metric].max()))
    leg = plt.legend()
    plt.xlabel('date')
    plt.ylabel('stock price')
```

```python
interact(plot, ticker=list(set(stocks['symbol'])), metric=['price', 'returns']);
```

```python
def plot_new(ticker='YHOO', price_metric='price', volume_metric='vol'):
    pyplot.show()
    fig = plt.figure(figsize=(15,9))
    data = stocks[stocks['symbol'] == ticker]
    data = data.sort_values(by=['date'], ascending=True)
    data.index = pd.DatetimeIndex(data['date'], name='date_')
    data['vol_sum'] = data['volume'].cumsum()
    data = data.resample('M').first()
    data['returns'] = data['price']/data['price'].shift(1)
    data['vol'] = data['vol_sum'] - data['vol_sum'].shift(1)
    data['% change'] = (data['vol']/data['vol'].shift(1))
    data['year'] = data['date'].apply(lambda x: x.year)
    data['month'] = data['date'].apply(lambda x: x.month)
    data['d_key'] = data['symbol'].astype(str)+data['year'].astype(str)+data['month'].astype(str)
    d = data.merge(cyber, left_on=data.d_key, right_on=cyber.key, how='left')
    d = d.fillna(0)
    d['data breach'] = np.where(d['Entity']!=0,d[price_metric],-100)
    d = d[['date', 'price', 'returns', 'symbol', 'vol', '% change', 'data breach', 'DATA SENSITIVITY']]
    d = d.iloc[2:]
   #plot for single y axis
   #plt.plot(d['date'], d[metric])
   #plt.scatter(d['date'].tolist(), d['data breach'], c=d['DATA SENSITIVITY'], s=d['DATA SENSITIVITY']*100, cmap = 'Reds')
   #plot for double y axis 
    ax=fig.add_subplot(111)
    ax2 = ax.twinx()
    ax.plot(d['date'], d[price_metric], linewidth=3.0)
    ax2.plot(d['date'], d[volume_metric], 'g-', alpha=0.3, linewidth=2.0)
   # plt.scatter(d['date'].tolist(), d['data breach'], c=d['DATA SENSITIVITY'], s=d['DATA SENSITIVITY']*100, cmap = 'Reds')
    ax.scatter(d['date'].tolist(), d['data breach'],c='Red', s=d['DATA SENSITIVITY']*100)
    #plt.ylim((0,d[metric].max()))
   # leg = plt.legend()
    ax.set_xlabel('date', fontsize=25)
    ax.set_ylabel(price_metric, color='b',fontsize=25 )
    ax2.set_ylabel(volume_metric, color='g',fontsize=25 )
    #plt.xlabel('date')
    #plt.ylabel('stock price')
```

```python
interact(plot_new, ticker=list(set(stocks['symbol'])), price_metric=['price', 'returns'], volume_metric=['vol','% change']);
```

```python
stock_dict = {
    i:stocks[stocks['symbol']==i] for i in set(stocks['symbol'])
}
```

```python
stock_dict['YHOO'].head()
```

```python
cyber['c_id'] = cyber['key']
```

```python
final_df=pd.DataFrame()
for i, _ in stock_dict.items():
    data = stock_dict[i]
    data['date'] = pd.to_datetime(data.date)
    data = data.sort_values(by=['date'], ascending=True)
    data.index = pd.DatetimeIndex(data['date'], name='date_')
    data = data.resample('M').first()
    data['returns'] = data['price']/data['price'].shift(1)
    data['year'] = data['date'].apply(lambda x: x.year)
    data['month'] = data['date'].apply(lambda x: x.month)
    data['d_key'] = data['symbol'].astype(str)+data['year'].astype(str)+data['month'].astype(str)
    d=pd.DataFrame()
    d = data.merge(cyber, left_on=data.d_key, right_on=cyber.c_id, how='left')
    d = d.fillna(0)
    d['month_before'] = np.where(d['Entity']!=0,d['returns'].shift(1),-100)
    d['month_breach'] = np.where(d['Entity']!=0,d['returns'],-100)
    d['month_after'] = np.where(d['Entity']!=0,d['returns'].shift(-1),-100)
    d = d[['date', 'price', 'returns', 'symbol', 'volume', 'month_before', 'month_breach', 'month_after']]
    d = d.iloc[1:]
    output = d[d['month_before']>0]
   # print(output.head())
    final_df = final_df.append(output)
    d=pd.DataFrame()
```
